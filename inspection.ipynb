```python
import numpy as np
import cv2
from utils import bbox2segm
from matplotlib import pyplot as plt
```

```python
MODEL_INPUT_SHAPE = (320, 320, 3)
```

```python
n = 0
ams_data = np.load('data/ams_data.npy')
ams_data = ams_data.reshape(ams_data.shape[0], *MODEL_INPUT_SHAPE)
ams_masks = bbox2segm(ams_data, MODEL_INPUT_SHAPE, (40, 40), 1)
ams_data_n_resized = cv2.resize(ams_data[n],(40,40))

fig, (ax1,ax2, ax3) = plt.subplots(1,3, figsize=(15,15))
ax1.imshow(ams_data[n])
ax2.imshow(ams_data_n_resized)
ax3.imshow(ams_masks[n])
```
